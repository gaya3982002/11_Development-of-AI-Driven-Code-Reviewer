Create Table for AI Code Reviewer

```python
import sqlite3
conn = sqlite3.connect('example.db')
cursor = conn.cursor()
```

```python
CREATE TABLE code_submissions (
    submission_id INT PRIMARY KEY AUTO_INCREMENT,
    student_name VARCHAR(50),
    code TEXT,
    submission_date DATETIME
);
```

Store Feedback

```python
CREATE TABLE ai_feedback (
    feedback_id INT PRIMARY KEY AUTO_INCREMENT,
    submission_id INT,
    error_count INT,
    style_score INT,
    ai_suggestions TEXT,
    FOREIGN KEY (submission_id) REFERENCES code_submissions(submission_id)
);
```

Review Output

```python
INSERT INTO ai_feedback (submission_id, error_count, style_score, ai_suggestions)
VALUES (1, 0, 85, 'Add docstring and improve variable names');
```

Fetch code analysis

```python
SELECT code
FROM code_submissions
WHERE submission_id = 1;
```

Error count

```python
SELECT SUM(error_count) AS total_errors
FROM ai_feedback;
```

Feedback report

```python
SELECT
    cs.student_name,
    af.error_count,
    af.style_score,
    af.ai_suggestions
FROM code_submissions cs
JOIN ai_feedback af
ON cs.submission_id = af.submission_id;
```
