Prediction

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])

model = LinearRegression()
model.fit(X, y)

print("Prediction for 6:", model.predict([[6]]))
```

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b

fibonacci(5)

```

Machine Learning -Linear Regression

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[10], [20], [30]])
y = np.array([1, 2, 3])   # complexity score

model = LinearRegression()
model.fit(X, y)

print("Predicted complexity:", model.predict([[25]]))
```

Chatbot logic

```python
def ai_explain(error):
    explanations = {
        "syntax": "Syntax error occurs due to incorrect code structure",
        "indentation": "Indentation error happens when spaces are inconsistent"
    }
    return explanations.get(error, "Unknown error")

print(ai_explain("indentation"))
```
